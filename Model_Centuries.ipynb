```python
import pandas as pd
df=pd.read_csv("/kaggle/input/sachin-tendulkar-100-centuries/Sachin Tendulkar - 100 Centuries.csv")
```

```python
print(df.head())
df.shape
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,6))
sns.histplot(df,x='Result',color='#ccd5ae', shrink = 0.9).set(xlabel = None)
plt.title("Result Vs Centuries")
plt.xlabel("Result")
plt.ylabel("No. of Centuries")
plt.tight_layout()
```

## 50 OUT OF 100 MATCHES ARE WON BY INDIA WHEN SACHIN SIR HITS A CENTURY

```python
plt.figure(figsize=(8,6))
sns.histplot(df,x='Result',hue='Captain',color='#ccd5ae', shrink = 0.9).set(xlabel = None)
plt.title("Result Vs Centuries")
plt.xlabel("Result")
plt.ylabel("No. of Centuries")
plt.tight_layout()
```

## VERY LESS CENTURIES AS A CAPTAIN

```python
sns.barplot(x=df['Position'],y=df['Strike Rate'])
```

## HIGHEST STRIKE RATE AT POSITION FOURTH

```python
plt.figure(figsize=(8,6))
sns.histplot(df,x='H/A',color='#ccd5ae', shrink = 0.9).set(xlabel = None)
plt.title("HOME OR AWAY")
plt.xlabel("Place")
plt.ylabel("No. of Centuries")
plt.tight_layout()
```

## ALMOST THE SAME NO. OF CENTURIES AT INDIA AND AWAY FROM INDIA


```python
plt.hist(df["Against"])
plt.xticks(rotation=90)
```

## MOST OF THE CENTURIES ARE AGAINST AUSTRALIA AND SOUTH AFRICA

```python
plt.figure(figsize = (8,6))
sns.histplot(df, x = 'Venue', color = '#ccd5ae', shrink = 0.9).set(xlabel = None)
plt.title('Number of Centuries by Venue')
plt.xticks(rotation=90)
plt.tight_layout()
```

## HE LIKES TO PLAY AT SHARJAH CRICKET STADIUM

```python
plt.figure(figsize=(8,6))
sns.histplot(df,x='Player of the match',hue='Result',color='#ccd5ae', shrink = 0.9).set(xlabel = None)
plt.title("Result Vs Centuries")
plt.xlabel("Player of the Match")
plt.ylabel("No. of Centuries")
plt.tight_layout()
```

## HIS MOST "THE PLAYER OF THE MATCH" ARE THE MATCHES WHEN INDIA LOSSES THE MATCH.

```python
plt.figure(figsize=(8,6))
sns.histplot(df,x='Innings',color='#ccd5ae', shrink = 0.9).set(xlabel = None)
plt.title("Inning Vs Centuries")
plt.xlabel("Inning No.")
plt.ylabel("No. of Centuries")
plt.xticks(range(1,5))
plt.tight_layout()
```

# SACHIN SIR MOSTLY SCORED IN THE FIRST INNING OF THE MATCH
